# Polynomial Regression

## Imports

```
import numpy as np
import operator
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error
from sklearn.preprocessing import PolynomialFeatures
```

## Creating the dataset

```
np.random.seed(0)
x = 2 - 3 * np.random.normal(0, 1, 20)
y = x - 2 * (x ** 2) + 0.5 * (x ** 3) + np.random.normal(-3, 3, 20)
```

## Transforming the dataset

```
x = x[:, np.newaxis]
y = y[:, np.newaxis]
```

## Creating a linear regression model and making a prediction

```
model = LinearRegression()
model.fit(x, y)
y_pred = model.predict(x)
```

## Checking the prediction

```
print('RMSE score of the linear regression is {}'.format(np.sqrt(mean_squared_error(y, y_pred)))) # 15.908242501429998
print('R2 score of linear regression is {}'.format(r2_score(y, y_pred))) # 0.6386750054827146
```

## Plotting the prediction

```
plt.scatter(x, y, s=10)
plt.plot(x, y_pred, color='r')
plt.show()
```

## Creating polynomial features

```
polynomial_features= PolynomialFeatures(degree=3)
x_poly = polynomial_features.fit_transform(x)
```

## Creating polynomial regression model and making a prediction

```
model = LinearRegression()
model.fit(x_poly, y)
y_poly_pred = model.predict(x_poly)
```

## Testing the prediction

```
print('RMSE score of the polynomial regression is {}'.format(np.sqrt(mean_squared_error(y, y_poly_pred)))) # 15.908242501429998
print('R2 score of linear regression is {}'.format(r2_score(y, y_poly_pred))) # 0.6386750054827146
```

## Plotting the prediction

```
plt.scatter(x, y, s=10)
# sort the values of x before line plot
sort_axis = operator.itemgetter(0)
sorted_zip = sorted(zip(x,y_poly_pred), key=sort_axis)
x, y_poly_pred = zip(*sorted_zip)
plt.plot(x, y_poly_pred, color='r')
plt.show()
```
